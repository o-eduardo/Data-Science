subtabela (linha com valores únicos)

### Tratando valores faltantes com pipline

```
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```
#preencher com valores de mediana
mediana = Pipeline(steps=[
                          ('mediana', SimpleImputer(strategy='median'))
])
```

```
#preenche com valores mais frequentes
frequente = Pipeline(steps=[
                          ('frequente', SimpleImputer(strategy='most_frequent', missing_values='?'))
])
```

### Criação de pipeline que compoe outros pipelines

```
colunas_valores_faltantes = list(set(colunas_valores_faltantes))
colunas_valores_faltantes
```

```
data_cleaning_ocup = ColumnTransformer(transformers=[
                                                ('frequent', frequente, ['occupation'])
])

data_cleaning_nativ = ColumnTransformer(transformers=[
                                                ('frequent', frequente, ['native-country'])
                                                
])

data_cleaning_wrkcl = ColumnTransformer(transformers=[
                                                ('frequent', frequente, ['workclass'])
                                                
])
```

### Criando pipeline final com o pre-processamento automatizado

>analisamos as colunas com valores faltantes preenchidas com '?'

>>Colunas: 'occupation', 'native-country', 'workclass'

> como são colunas com valores de string (object), vamos usar o preenchimento com valores mais frequentes



```
colunas_valores_faltantes
```

```
pipeline_final = Pipeline([
                           ('datacleaningocup', data_cleaning_ocup),
                           #('data_cleaningnativ', data_cleaning_nativ),
                           #('data_cleaningwork', data_cleaning_wrkcl),
                           ('ohe', OneHotEncoder()),
                           ('standartscaler', StandardScaler()),
                           ('tree', tree.DecisionTreeClassifier())
])
```

```
pipeline_final.steps
```
